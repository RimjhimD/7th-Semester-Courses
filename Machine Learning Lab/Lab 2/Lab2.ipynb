```python
print("Missing values per column:")
df.isnull().sum()
```

 Analyze Numerical Column Statistics


```python
numerical_columns = ['age', 'study_hours', 'class_attendance', 'sleep_hours', 'exam_score']
print("Descriptive statistics for numerical columns:")
df[numerical_columns].describe()
```

 Analyze Numerical Column Statistics and Outliers



```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.show()
```

Inspect Categorical Column Values for Inconsistencies

```python
print("Unique values and their counts for categorical columns:")
for col in categorical_columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```



### Data Analysis Key Findings

*   **Numerical Data Quality**: The analysis of numerical columns (`age`, `study_hours`, `class_attendance`, `sleep_hours`, `exam_score`) using box plots revealed a lack of significant outliers. This indicates that these columns have well-distributed, consistent, and realistic ranges, suggesting high data integrity for these quantitative measures.
*   **Categorical Data Consistency**: All categorical columns (`gender`, `course`, `internet_access`, `sleep_quality`, `study_method`, `facility_rating`, `exam_difficulty`) were found to be highly consistent. Unique value counts confirmed no unexpected entries, misspellings, or inconsistencies, with entries adhering to predefined categories (e.g., `internet_access` only containing 'yes' and 'no'; `sleep_quality` containing 'poor', 'good', 'average').
*   **Overall Dataset Quality**: The dataset is deemed to be of high quality, with no missing values (as confirmed in prior checks), and consistent, well-defined entries across both numerical and categorical attributes. The absence of significant outliers and the clean categorization suggest the data is reliable for direct use without extensive pre-processing or imputation.
